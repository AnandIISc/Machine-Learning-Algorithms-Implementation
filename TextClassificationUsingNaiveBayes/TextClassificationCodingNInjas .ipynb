```python
# handeling directory and file with file operator
X  =[] 
Y = []
for category in os.listdir("D:\\Machine Learning by coding Ninja\\8.Naive Bayes\\20_newsgroups"):
    for document in os.listdir("D:\\Machine Learning by coding Ninja\\8.Naive Bayes\\20_newsgroups\\"+category):
        with open("D:\\Machine Learning by coding Ninja\\8.Naive Bayes\\20_newsgroups\\"+category+'\\'+document, "r") as f:
            X.append((document,f.read()))
            Y.append(category)
```

```python
print(type(X))
print(type(X[0]))
print(len(Y))
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y)
```

```python
dic={}
for i in range(len(x_train)):
    word=x_train[i][1].lower()
    stripped=re.split(r'\W+',word)              #spliting word with help of re.split
    for j in stripped:
        #removing stop word and cheacking character with .isalpha 
        if not(j.isalpha()) or j in stop_word:
            continue
        if j in dic:
            dic[j] += 1 
        else:
            dic[j] = 1
dic
```

```python
sorted_dic = sorted(dic.items(), key=operator.itemgetter(1),reverse=True)
sorted_dic
```

```python
len(sorted_dic)
features=sorted_dic
```

```python
#making our model to 5000 words
answer1=[features[i][0] for i in range(5000)]
answer1
```

```python
x_train_dataset=np.zeros([len(x_train),len(answer1)],int)
for i in range(len(x_train)):
    data=x_train[i][1].lower()
    word=re.split(r'\W+',data)
    #Iterating over each word
    for j in word:
        #We will add the frequency corresponding to that word only which is in our answer1(feature list)
        if j in answer1:
            x_train_dataset[i][answer1.index(j)]+=1
```

```python
x_test_dataset=np.zeros([len(x_test),len(answer1)],int)
for i in range(len(x_test)):
    data=x_test[i][1].lower()
    word=re.split(r'\W+',data)
    for j in word:
        if j in answer1:
            x_test_dataset[i][answer1.index(j)]+=1
```

```python
def fit(x_train_dataset,y_train):
    count={}
    total_word=0
    y_train=np.array(y_train)
    count["total_doc"]=len(y_train)
    classes=set(y_train)
    for i in classes:
        temp=0
        x_train_with_i=x_train_dataset[y_train==i]
        temp2=x_train_with_i.shape[0]
        count[i]={}
        for feature in answer1:
            l=(x_train_with_i[:,answer1.index(feature)]).sum()
            count[i][feature]=l
            temp+=l
        count[i]["word_in_class"]=temp
        count[i]["length"]=temp2
        
    
    return count
```

```python
def probability(x_test,dic,classes):
    prob=np.log(dic[classes]["length"])-np.log(dic["total_doc"])
    feature=list(dic[classes].keys())
    #-2 is done becuase there will be "length" and "word in class" present in feature. 
    for j in range (len(feature)-2):
        xj=x_test[j]
        #If frequency is 0, we will not consider it
        if xj==0:
            current_prob=0
        else:
            #Extra addition part is Laplace correction
            num=dic[classes][feature[j]]+1
            den=dic[classes]["word_in_class"]+len(dic[classes].keys())-2
            current_prob=np.log(num)-np.log(den)
        prob+=current_prob
    return prob
```

```python
def predict_for_single(x_test,dic):
    first_run=True
    classes=dic.keys()
    for i in classes:
        if i=="total_doc":
            continue
        prob=probability(x_test,dic,i)
        if first_run or prob>best_prob:
            best_prob=prob
            first_run=False
            best_class=i
    return best_class
```

```python
def predict_(x_test,dic):
    y_pred=[]
    for x in x_test:
        y_pred.append(predict_for_single(x,dic))
    return y_pred
```

```python
def score(y_test,y_pred):
        count = 0
        for i in range(len(y_pred)):
            if y_pred[i] == y_test[i]:
                count+=1
        return count/len(y_pred)
```

```python
dictionary=fit(x_train_dataset,y_train)
y_pred=predict_(x_test_dataset,dictionary)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print("Score on testing with Implemented Function:",score(y_test,y_pred))
```
